```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("cancer_data.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.heatmap(data.isnull())
```

```python
data.drop(["Unnamed: 32", "id"], axis = 1, inplace=True)
data.head() 
```

```python
data.diagnosis = [1 if value == "M" else 0 for value in data.diagnosis]
data.head()
```

```python
#splitting into target vs predictor variable 
y = data["diagnosis"]
x = data.drop(["diagnosis"], axis=1)
y
```

## Normalization of data




```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_scaled = scaler.fit_transform(x)

X_scaled
```

```python
#Splitting the data to train and testing 

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=42)

#Train model 
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

#fit the model on training data 
lr.fit(X_train, Y_train)

#predict the target variable on test data 
y_pred = lr.predict(X_test)
```

```python
y_pred # 0's and 1's show if there is cancer or not 
```

```python
Y_test
```

## Evaluation of the model 


```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y_test, y_pred)
print(f"Accuracy: {accuracy: .2f}")
```

```python
from sklearn.metrics import classification_report
print (classification_report(Y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt 


cm = confusion_matrix(Y_test, y_pred)

plt.figure(figsize=(8, 4))
sns.heatmap(cm, cmap='Reds', annot=True, fmt='d')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
plt.scatter(Y_test, y_pred, color='blue', label='Actual Data')
plt.title('Breast Cancer detection')
plt.xlabel('Feature')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
